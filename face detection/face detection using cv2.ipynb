```python
import numpy as np
import cv2
import os

def add_zero(count):
    if 9 // i:
        return '0000'+str(i)
    elif 99 // i:
        return '000'+str(i)
    elif 999 // i:
        return '00'+str(i)
    elif 9999 // i:
        return '0'+str(i)
    elif 99999 // i:
        return str(i)
    
ROOT_DIR = os.getcwd()
DATA_DIR = os.path.join(ROOT_DIR, 'Data')
AFLW_DATA_DIR = os.path.join(DATA_DIR, 'aflw')
SAVE_DIR = os.path.join(DATA_DIR, 'face')

face_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.3.1_1/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml')

if not face_cascade.empty():
    print('face_cascade read success!!')
    
count = 1
percentage = 0.01
MAX_IMAGE_COUNT = 71362

print('face detection start!!')
print('======================')

for i in range(1,MAX_IMAGE_COUNT):
    try:
        img = cv2.imread(AFLW_DATA_DIR+'/image'+add_zero(i)+'.jpg')
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x,y,w,h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]

            img_resize = cv2.resize(roi_color, (128,128))
            img_resize_gray = cv2.resize(roi_gray, (128,128))
            
            cv2.imwrite(SAVE_DIR +'/'+str(count)+'.jpg', img_resize)
            cv2.imwrite(SAVE_DIR +'/gray_'+str(count)+'.jpg', img_resize_gray)
            
            count += 1
            
        cv2.destroyAllWindows()

        if int(MAX_IMAGE_COUNT * percentage) < i:
            print(int(percentage * 100) , '% complete')
            percentage += 0.01

    except:
        continue
        #print('don\'t have image'+ add_zero(i) +'.jpg')

print('======================')
print('finish!!')
```
